```python
aa = torch.tensor([[float(j+i) for i in range(3)] for j in range(4)])
```

```python
aa
```

```python
torch.linalg.norm(aa, dim=-1)
```

```python
torch.sqrt(torch.matmul(aa, aa.T))
```

```python
bb = torch.div(aa, torch.linalg.norm(aa, dim=-1)[:, None])
```

```python
torch.sqrt(torch.matmul(bb, bb.T))
```

```python
aa_new = torch.rand(64, 3)
```

```python
aa_new
```

```python
bb_new = torch.div(aa_new, torch.linalg.norm(aa_new, dim=-1)[:, None])
```

```python
bb_new.shape, bb_new, torch.diagonal(torch.sqrt(torch.matmul(bb_new, bb_new.T)))
```

```python
bb_new[0:16, :].shape, 
```

```python
sim_mat = torch.matmul(bb_new, bb_new.T)
for i, row in enumerate(sim_mat):
    print(i, row)
```

```python
sim_mat[0:16, 0:16].shape, sim_mat[0:16, 16:32].shape, sim_mat[0:16, 32:].shape
```

```python
posv  = torch.diagonal(     sim_mat[0:16, 16:32]  )
negv  = torch.multinomial(  sim_mat[0:16, 32:],  3)
print(posv.shape, negv.shape)
```
